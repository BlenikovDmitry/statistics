```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.figure import Figure
import statistics
import csv
import generator_muni as gen

data = pd.read_csv('muni.csv', encoding='Windows-1251', decimal=',')
#    ,   _,     
data.columns = data.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('.', '')

#   :
#   
data[''] = data[''].str.lower()
data[''] = data[''].str.lower()
data['_'] = data['_'].str.lower()
data['_'] = data['_'].str.lower()

#       ,   0
data['_'] = data['_'].astype(str).str.replace('-', '0')
#  float  ,  ,    _
data['_'] = data['_'].str.replace(',', '.').astype(float)
data['_'] = data['_'].astype(float)
#  _
data['_'] = round(((data['_'].astype(float) / data['_'].astype(float) * 4) * 100), 2)

#  ,     
data['_%_'] = data['_%_'].astype(str).str.replace('-', '0')
data['_%_'] = data['_%_'].str.replace(',', '.').astype(float)
data = data.loc[data['_%_'].astype(float) > 0]

#         
# ,         2-  
# ,        2-  
data['__'] = data['__'].astype(str).str.replace(',', '.').astype(float)

stde_interest = statistics.stdev(data['_'])
average_interest = statistics.mean(data['_'])
stde_doh_last = statistics.stdev(data['__'])
average_doh_last = statistics.mean(data['__'])
stde_price = statistics.stdev(data['_%_'])
average_price = statistics.mean(data['_%_'])
data = data.loc[data['__'].astype(float) <= ((stde_doh_last * 2) + average_doh_last)]
data = data.loc[data['__'].astype(float) > 0]
data = data.loc[data['_%_'].astype(float) <= ((stde_price * 2) + average_price)]
data = data.loc[data['_%_'].astype(float) > 0]

stde_interest = statistics.stdev(data['_'])
average_interest = statistics.mean(data['_'])
stde_doh_last = statistics.stdev(data['__'])
average_doh_last = statistics.mean(data['__'])
stde_price = statistics.stdev(data['_%_'])
average_price = statistics.mean(data['_%_'])
sum_debt = data['__'].astype(float).sum(axis = 0)


#   ,  ,   ,    ,    , 
#  ,    
#      ,          
#  -  ,        
'''with open('result_muni.csv', 'a') as f:
    writer = csv.writer(f,  lineterminator='\n')                   
    writer.writerow([str(data['__'][0])] + [str(round(average_interest, 2))] + [str(round(stde_interest, 2))] +
                   [str(round(average_doh_last, 2))] + [str(round(stde_doh_last, 2))] + [str(round(average_price, 2))] +
                   [str(round(stde_price, 2))] + [str(sum_debt)] )
#     .        
data_tmp = pd.read_csv('result_muni.csv', encoding='Windows-1251')
data_tmp = data_tmp.drop_duplicates()
data_tmp.to_csv('result_muni.csv', index=False)'''

data.to_csv('result/muni_cleared.csv')
#   
fig = plt.figure(figsize=(25,10))
ax = fig.add_subplot(1,1,1)
plt.xticks(rotation="vertical")
plt.yticks(np.arange(min(data['_']), max(data['_'])+1, 1.0))
ax.set_title(" ")
ax.plot(data[''],data['_'], '-rh', linewidth=3, markersize=5, markerfacecolor='b')
ax.grid()
plt.savefig('result/kupon_muni.png')
#   
fig = plt.figure(figsize=(25,10))
ax = fig.add_subplot(1,1,1)
plt.xticks(rotation="vertical")
plt.yticks(np.arange(min(data['__']), max(data['__'])+1, 1.0))
ax.set_title("    ")
ax.plot(data[''],data['__'], '-gh', linewidth=3, markersize=5, markerfacecolor='b')
ax.grid()
plt.savefig('result/dohod_eff_muni.png')
#   
fig = plt.figure(figsize=(25,10))
ax = fig.add_subplot(1,1,1)
plt.xticks(rotation="vertical")
plt.yticks(np.arange(min(data['_%_']), max(data['_%_'])+1, 1.0))
ax.set_title(" ")
ax.plot(data[''],data['_%_'], '-bh', linewidth=3, markersize=5, markerfacecolor='b')
ax.grid()
plt.savefig('result/price_muni.png')
#  
fig = plt.figure(figsize=(25,10))
ax = fig.add_subplot(1,1,1)
plt.xticks(rotation="vertical")
ax.set_title(" ")
ax.hist(data['_'].astype(int),16,linewidth = 3, color = "lightblue", ec="red")
ax.grid()
plt.savefig('result/interest_distribution_muni.png')

#      
data_tmp = data.sort_values(by='_%_', ascending=False)
#     
fig = plt.figure(figsize=(25,10))
ax = fig.add_subplot(1,1,1)
plt.xticks(rotation="vertical")
plt.yticks(np.arange(min(data_tmp['_%_']), max(data_tmp['_%_'])+1, 1.0))
ax.set_title(" ")
ax.plot(data_tmp[''],data_tmp['_%_'], '-bh', linewidth=3, markersize=5, markerfacecolor='b')
ax.grid()
plt.savefig('result/price_muni_sorted.png')
#       
data_tmp = data.sort_values(by='_', ascending=False)
#    
fig = plt.figure(figsize=(25,10))
ax = fig.add_subplot(1,1,1)
plt.xticks(rotation="vertical")
plt.yticks(np.arange(min(data_tmp['_']), max(data_tmp['_'])+1, 1.0))
ax.set_title(" ")
ax.plot(data_tmp[''],data_tmp['_'], '-rh', linewidth=3, markersize=5, markerfacecolor='b')
ax.grid()
plt.savefig('result/kupon_muni_sorted.png')

#     
data_tmp = data
data_tmp['_'] = pd.to_datetime(data_tmp['_'], format='mixed')
data_tmp = data.sort_values(by='_', ascending=False)
#    
fig = plt.figure(figsize=(25,10))
ax = fig.add_subplot(1,1,1)
plt.xticks(rotation="vertical")
ax.set_title(" ")
ax.plot(data_tmp[''],data_tmp['_'], '-bh', linewidth=3, markersize=5, markerfacecolor='b')
ax.grid()
plt.savefig('result/end_graphic_muni.png')

##########################################################3
#      
data_tmp = data
data_tmp = data.sort_values(by='__', ascending=False)
#    
fig = plt.figure(figsize=(25,10))
ax = fig.add_subplot(1,1,1)
plt.xticks(rotation="vertical")
ax.set_title("  ")
ax.plot(data_tmp[''],data_tmp['__'], '-gh', linewidth=3, markersize=5, markerfacecolor='b')
ax.grid()
plt.savefig('result/dohod_eff_muni_sorted.png')
############################################################################################
#      __
data_tmp = data
data_tmp['__'] = data_tmp['__'].astype(int)
data_tmp = data.sort_values(by='__', ascending=False)
fig = plt.figure(figsize=(25,10))
ax = fig.add_subplot(1,1,1)
plt.xticks(rotation="vertical")
ax.set_title(" ")
ax.bar(data_tmp[''], data_tmp['__'], width=1, color = "lightblue", edgecolor="red", linewidth=0.7)
ax.grid()
plt.savefig('result/volume_muni_sorted.png')

arg = data['__'].values[:1].astype(str)
arg = str(arg)
arg = arg[2:10]
gen.init(arg)

rows = data.columns.tolist()
gen.table(rows)

counter = 0
while counter < len(data):
    data_tmp = data.iloc[counter]
    rows = data_tmp.values.tolist()
    gen.table(rows)
    counter+=1
gen.statistic_header()
rows = [str(round(average_interest, 2)) + "%", str(round(stde_interest,2))+ "%",str(round(average_doh_last,2))+ "%",
                                             str(round(stde_doh_last,2))+ "%", str(round(average_price,2))+ "%",str(round(stde_price,2))+ "%",
                                             str(round(sum_debt,2))]
gen.statistic_body(rows)
gen.graphics()



```
